```python
x = var('x')
p = var('p')
f = x^3 - 3*x^2 - 2*x + 1; f
```

```python
approx = f(x=0); approx
```

```python
sum([plot(func, (v, -2, 4), color=c) for func, v, c in [(f, x, '#0000FF'), (approx, p, '#00FF00')]])
```

```python
df = diff(f)
approx = f(x=0) + df(x=0)*p; approx
```

```python
df
```

```python
sum([plot(func, (v, -2, 4), color=c) for func, v, c in [(f, x, '#0000FF'), (approx, p, '#00FF00'), (df, x, '#ee2222')]])
```

```python
d2f = diff(df)
approx = f(x=0) + df(x=0)*p + p^2/factorial(2)*d2f(x=0); approx
```

```python
d2f
```

```python
sum([plot(func, (v, -2, 4), color=c) for func, v, c in [(f, x, '#0000FF'), (approx, p, '#00FF00'), (df, x, '#ee2222'), (d2f, x, '#996666')]])
```

```python
d3f = diff(d2f)
approx = f(x=0) + df(x=0)*p + p^2/factorial(2)*d2f(x=0) + p^3/factorial(3)*d3f(x=0)
```

```python
d3f
```

```python
sum([plot(func, (v, -2, 4), color=c) for func, v, c in [(f, x, '#0000FF'), (approx, p, '#00FF00'), (df, x, '#ee2222'), (d2f, x, '#996666'), (d3f, x, '#66eeee')]])
```
