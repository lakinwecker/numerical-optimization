```python
x = var('x')
f = x^4 * cos(1/x) + 2*x^4
solve(f, x)
```

```python
plot(f, (x, -0.001, 0.001))
```

```python
df = diff(f); df
```

```python
plot(df, (x, -0.001, 0.001))
```

```python
df2 = diff(df); df2
```

```python
plot(df2, (x, -0.001, 0.001))
```

```python
df3 = diff(df2); df3
```

```python
plot(df3, (x, -0.001, 0.001))
```
